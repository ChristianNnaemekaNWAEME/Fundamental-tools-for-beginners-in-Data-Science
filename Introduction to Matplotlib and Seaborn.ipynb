**Matplotlib is useful for making Plots**

```python
# importing matplotlib Library
import matplotlib.pyplot as plt
# import numpy to get data for our plots
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

Plotting the data

```python
# sine wave
plt.plot(x,y)
plt.show
```

```python
# Cosine wave
plt.plot(x,z)
plt.show
```

```python
# adding title in x and y-axis labels
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
```

```python
# Parabola
x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
# Parabola
x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y, 'rx')
plt.show()
```

```python
x = np.linspace(-5,5,50)
plt.plot(x, np.sin(x), 'g')
plt.plot(x, np.cos(x), 'r--')
plt.show
```

**Bar Plot**

```python
fig = plt.figure()
ax =fig.add_axes([0,0,1,1])
languages = ['English', 'French', 'Spanish', 'Latin', 'German']
people = [100, 50, 150, 40, 70]
ax.bar(languages, people)
plt.xlabel('LANGUAGES')
plt.ylabel('NUMBER OF PEOPLE')
plt.show()
```

Pie chart

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['English', 'French', 'Spanish', 'Latin', 'German']
people = [100, 50, 150, 40, 70]
ax.pie(people, labels=languages, autopct = '%1.1f%%')
plt.show()
```

Scatter Plot

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='r')
ax.scatter(x,z, color = 'b')
plt.show()
```

3D scatter Plot

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap = 'Blues')
plt.show()
```

Seaborn has some built-in datasets

```python
# Total bills vs tip dataset
tips =sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#tips.isnull().sum()
```

```python
# setting a theme for the plots
sns.set_theme()
```

```python
# Visualize the data
sns.relplot(data=tips, x = 'total_bill', y = 'tip', col = 'time', hue = 'smoker', style = 'smoker', size = 'size')
```

```python
# House price dataset
from sklearn.datasets import load_boston
house_boston = load_boston()
house = pd.DataFrame(house_boston.data, columns=house_boston.feature_names)
```

```python
house.head()
```

```python
house['PRICE'] = house_boston.target
```

```python
house.head()
```

**Distribution plot**

```python
sns.distplot(house['PRICE'])
```

**Correlation Matrix:**
    
    1. Positive
    2. Negative

Heat Map

```python
correlation = house.corr()
# Constructing a Heat Map
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar =True, square = True, fmt = '.1f', annot = True, annot_kws = {'size':8}, cmap = 'Blues')
```
