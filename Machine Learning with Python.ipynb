**Find most recent & the earliest traveled date and mean distance traveled for each start city**

```python
df = df.iloc[:-1,:]
df.tail()
df['START_DATE*'] = pd.to_datetime(df['START_DATE*'])
df.dtypes 
```

```python
# Solution
res = df.groupby('START*').agg({'MILES*':'mean',
                                'START_DATE*':['min','max']})
res
res.reset_index(inplace=True)

res
res.columns=['START_CITY','AVG_DIST','EARIEST_TRIP', 'RECENT_TRIP']
res.to_csv("raghu.csv")
```
